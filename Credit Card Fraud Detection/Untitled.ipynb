```python
# Correlation before data correction
corr = df.corr()
plt.figure(figsize=(25, 25))
sns.heatmap(round(corr, 3), annot=True)
plt.title("Correlation before outlier removal");
```

* No correlation or very less correlation

```python
def remove_outliers(df, feature):
    
    IQR = df[feature].quantile(0.75) - df[feature].quantile(0.25)  # Inter Quartile Range
    lower_bound = df[feature].quantile(0.25) - 1.5*IQR
    upper_bound = df[feature].quantile(0.75) + 1.5*IQR

    df[feature] =  np.where(df[feature] <= lower_bound, lower_bound,  df[feature])
    df[feature] =  np.where(df[feature] >= upper_bound, upper_bound,  df[feature])
```

```python
df_clean = df.copy()

for feature in features:
    remove_outliers(df_clean, feature)

df_clean.head()
```

```python
%%time
# Checking for outliers
fig, axes = plt.subplots(6, 5, figsize=(15, 20))
axes = axes.flatten()

for ax, feature in zip(axes, features):
    sns.boxplot(x = df_clean[feature], ax=ax)
```

```python
# Taking every independent feature
X = df_clean.iloc[:, :-1]
Y = df_clean.iloc[:, -1]

print("X shape", X.shape)
print("y shape", Y.shape)
```

## Train/Test Split

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
```

```python
# train test split
X_train ,X_test, Y_train ,Y_test =  train_test_split(X, Y,  test_size=0.25, random_state=0, stratify=Y)
```

```python
# Standard Scaling Data
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

# ModelTraining

```python
from sklearn.linear_model import LogisticRegressionCV, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
def model_result(model):
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    train_score = model.score(X_train, Y_train)
    print("Model score {}".format(train_score))
    print("-" * 50)
    print()
    print('Classification Report:\n', classification_report(Y_test, y_pred))
    print("-" * 50)
    print()
    test_score = accuracy_score(Y_test, y_pred)
    print("Test accuracy Score: ", test_score)
    print("-" * 50)
    print()
    plt.figure(figsize=(10,10))
    sns.heatmap(confusion_matrix(Y_test, y_pred), annot=True, fmt='g');
    return train_score, test_score
```

```python
result = pd.DataFrame({"Models":[], "Training Acc":[], "Test Acc":[]})
result
```

```python
train_score, test_score = model_result(LogisticRegressionCV())
result.loc[len(result.index), :] = ["Logistic Regression", train_score, test_score]
```

```python
train_score, test_score = model_result(DecisionTreeClassifier(max_depth=3))
result.loc[len(result.index), :] = ["Decision Tree", train_score, test_score]
```

```python
train_score, test_score = model_result(SGDClassifier())
result.loc[len(result.index), :] = ["SGD classifier", train_score, test_score]
```

```python
train_score, test_score = model_result(SVC())
result.loc[len(result.index), :] = ["SVC", train_score, test_score]
```

```python
train_score, test_score = model_result(GaussianNB())
result.loc[len(result.index), :] = ["Gaussian Naive Base", train_score, test_score]
```

```python
train_score, test_score = model_result(RandomForestClassifier())
result.loc[len(result.index), :] = ["Random Forest", train_score, test_score]
```

```python
result.sort_values(by=['Test Acc', 'Training Acc'], ascending=False)
```
